```python
import pandas as pd
path=r'C:\Users\DEVA NANTHAN\Documents\youtube\youtube_ad_revenue_dataset.csv'
df=pd.read_csv(path)
df
```

```python
df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
df['comments']=df['comments'].fillna(0)
df['likes']=df['likes'].fillna(0)
df=df[df['ad_revenue_usd']>=0]
```

```python
df['watch_time_minutes']=df['watch_time_minutes'].fillna(df['watch_time_minutes'].median())
```

```python
import numpy as np
df['engagement_rate']=(df['likes']+df['comments'])/df['views'].replace(0, np.nan)
```

```python
df['watch_per_view']=df['watch_time_minutes']/df['views'].replace(0, np.nan)
df['date']=pd.to_datetime(df['date'])
```

```python
df['month']=df['date'].dt.month
df['is_weekend']=df['date'].dt.dayofweek.apply(lambda x: 1 if x in [5,6] else 0)
```

```python
encode={'Education':0, 'Tech':1, 'Music':2, 'Entertainment':3, 'Gaming':4, 'Lifestyle':5}
df['category']=df['category'].map(encode)
```

```python
encode={'TV':0, 'Desktop':1, 'Tablet':2, 'Mobile':3}
df['device']=df['device'].map(encode)
```

```python
encode={'CA':0, 'DE':1, 'IN':2, 'AU':3, 'UK':4, 'US':5}
df['country']=df['country'].map(encode)
```

```python
df
```

```python
x=df.drop(columns=['ad_revenue_usd','video_id','date','device','country'])
y=df['ad_revenue_usd']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=mse**0.5
r2=r2_score(y_test,y_pred)
```

```python
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R2: {r2}')
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1.0)
ridge.fit(x_train,y_train)
y_pred_ridge=ridge.predict(x_test)
mae_ridge=mean_absolute_error(y_test,y_pred_ridge)
mse_ridge=mean_squared_error(y_test,y_pred_ridge)
rmse_ridge=mse_ridge**0.5
r2_ridge=r2_score(y_test,y_pred_ridge)
print(f'Ridge MAE: {mae_ridge}')
print(f'Ridge MSE: {mse_ridge}')
print(f'Ridge RMSE: {rmse_ridge}')
print(f'Ridge R2: {r2_ridge}')
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.1)
lasso.fit(x_train,y_train)
y_pred_lasso=lasso.predict(x_test)
mae_lasso=mean_absolute_error(y_test,y_pred_lasso)
mse_lasso=mean_squared_error(y_test,y_pred_lasso)
rmse_lasso=mse_lasso**0.5
r2_lasso=r2_score(y_test,y_pred_lasso)
print(f'Lasso MAE: {mae_lasso}')
print(f'Lasso MSE: {mse_lasso}')
print(f'Lasso RMSE: {rmse_lasso}')
print(f'Lasso R2: {r2_lasso}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr.fit(x_train,y_train)
y_pred_gbr=gbr.predict(x_test)
mae_gbr=mean_absolute_error(y_test,y_pred_gbr)
mse_gbr=mean_squared_error(y_test,y_pred_gbr)
rmse_gbr=mse_gbr**0.5
r2_gbr=r2_score(y_test,y_pred_gbr)
print(f'GBR MAE: {mae_gbr}')
print(f'GBR MSE: {mse_gbr}')
print(f'GBR RMSE: {rmse_gbr}')
print(f'GBR R2: {r2_gbr}')
```

```python
import pickle
with open('scaler.pkl','wb') as f:
    pickle.dump(scaler,f)
with open('linear_model.pkl','wb') as f:
    pickle.dump(lr,f)
with open('ridge_model.pkl','wb') as f:
    pickle.dump(ridge,f)
with open('lasso_model.pkl','wb') as f:
    pickle.dump(lasso,f)
with open('gbr_model.pkl','wb') as f:
    pickle.dump(gbr,f)
```
